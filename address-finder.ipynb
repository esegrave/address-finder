```python
# import data
import pandas as pd
import numpy as np

df = pd.read_csv('organizations.csv')
```

```python
# filter to just org names
organizations = df['Organization Name']
```

```python
# initialize API key
API_KEY = <enter-key>

# create request path
API_PATH = 'https://maps.googleapis.com/maps/api/place/findplacefromtext/json'
```

```python
import requests

# initialize list of addresses
addresses = []

# loop through orgs, calling place search to get place info for each organzation
for org in organizations:
    payload = {'input': org, 'inputtype': 'textquery', 'fields' : 'formatted_address', 'key' : API_KEY}
    r = requests.get(API_PATH, params = payload)
    try:
        addresses.append(r.json()['candidates'][0]['formatted_address'])
    except:
        addresses.append('Cannot find address.')
```

```python
df['addresses'] = addresses
```

```python
df.to_csv('organizations-with-address.csv', index=False)
```
